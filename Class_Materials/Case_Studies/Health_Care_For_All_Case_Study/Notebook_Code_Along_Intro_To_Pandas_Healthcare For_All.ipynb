Adding a column "rich" for donors with a household income greater than 350

```python
rich = []

for i in range(len(health4all)):
    if(health4all["median_household_income"][i]>350):
        rich.append(True)
    else:
        rich.append(False)
```

```python
health4all["rich"] = rich
```

```python
health4all["rich"]= [True if health4all["median_household_income"][i]>350 else False for i in range(len(health4all))]
```
